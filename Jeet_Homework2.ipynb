## Problem 4

```r
# Experiment 1 

# simulations of a single die  
sim = ceiling(runif(10000, min = 0, max = 6)) 

a = 0
b = 6

dat4a = data.frame(length=sim)
ggplot(dat4a, aes(length)) + 
    geom_histogram(bins = 6, alpha = 1)
```

```r
# Experiment 2 

# simulations of 2 dice 
sim2 = ceiling(runif(10000, min = 0, max = 6)) + ceiling(runif(10000, min = 0, max = 6))

a = 0
b = 6

dat4b = data.frame(length=sim2)
ggplot(dat4b, aes(length)) + 
    geom_histogram(bins = 6, alpha = 1)
```

```r
# Experiment 3

# simulations of 3 dice 
sim3 = ceiling(runif(10000, min = 0, max = 6)) + ceiling(runif(10000, min = 0, max = 6)) + 
    ceiling(runif(10000, min = 0, max = 6))

a = 0
b = 6

dat4c = data.frame(length=sim3)
ggplot(dat4c, aes(length)) + 
    geom_histogram(bins = 6, alpha = 1)
```

```r
# Experiment 4

# simulations of 6 dice 
sim4 = ceiling(runif(10000, min = 0, max = 6)) + ceiling(runif(10000, min = 0, max = 6)) + 
    ceiling(runif(10000, min = 0, max = 6)) + ceiling(runif(10000, min = 0, max = 6)) +
    ceiling(runif(10000, min = 0, max = 6)) + ceiling(runif(10000, min = 0, max = 6))

a = 0
b = 6

dat4d = data.frame(length=sim4)
ggplot(dat4d, aes(length)) + 
    geom_histogram(bins = 6, alpha = 1)
```

```r
ggplot(dat4a, aes(sample=dat4a)) + geom_point(stat = "qq") + ggtitle("Experiment 1")
```

```r
ggplot(dat4b, aes(sample=dat4b)) + geom_point(stat = "qq") + ggtitle("Experiment 2")
```

```r
ggplot(dat4c, aes(sample=dat4c)) + geom_point(stat = "qq") + ggtitle("Experiment 3")
```

```r
ggplot(dat4d, aes(sample=dat4d)) + geom_point(stat = "qq") + ggtitle("Experiment 4")
```

From the above distributions and QQ-plots of the four different experiments, we notice that Experiment 4 which simulated the dice roll of 4 dice is the most similar to a normal distribution. This is evident from both the histogram as well as the QQ-plot. This result is consist with the central limit theorem, as experiment 4 has the maximum number of dice being rolled, and according to this theorem, that for larger sample sizes the x-bar is normally distributed.
