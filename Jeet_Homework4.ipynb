### 1. 
#### (a) 


$ \text{E}[(Y - (b_0 + b_1X))^2] = \text{E}[Y^2] +  \text{E}[(b_0 + b_1X)^2] + 2\text{E}[Y(b_0 + b_1X)]$ 


$\Rightarrow \text{E}[Y^2] +  \text{E}[b_0^2] + \text{E}[b_1^2X^2] + 2\text{E}[b_0b_1X] - 2b_0\text{E}[Y] - 2b_1\text{E}[XY]$
$\Rightarrow  \text{E}[Y^2] +  b_0^2 + b_1^2(\text{Var}[X] + \text{E}[X]^2) + 2b_0b_1\text{E}[X] - 2b_0\text{E}[Y] - 2b_1\text{Cov}[X,Y]  - 2b_1\text{E}[X]\text{E}[Y]$
$\Rightarrow \text{E}[Y^2] - 2b_0\text{E}[Y] - 2b_1\text{Cov}[X,Y] - 2b_1\text{E}[X]\text{E}[Y] + b_0^2 + b_1^2\text{Var}[X] + b_1^2(\text{E}[X])^2 + 2b_0b_1\text{E}[X]$ 

#### (b)

$\frac{\partial{\text{MSE}(b_0, b_1)}}{\partial{b_0}} = -2\text{E}[Y] + 2b_0 + 2b_1\text{E}[X]$

Setting derivative to 0 and solving for $b_0$, we get: 

$-2\text{E}[Y] + 2b_0 + 2b_1\text{E}[X] = 0$
$\Rightarrow b_0 + b_1\text{E}[X] - \text{E}[Y] = 0$
$\Rightarrow b_0 =  \text{E}[Y] - b_1\text{E}[X]$
$\square$

#### (c) 

$\frac{\partial{\text{MSE}(b_0, b_1)}}{\partial{b_1}} = -2\text{Cov}[X,Y] - 2\text{E}[X]\text{E}[Y] + 2b_0\text{E}[X] + 2b_1\text{Var}[X] + 2b_1(\text{E}[X])^2$

Setting derivative to 0 and solving for $b_1$, we get: 

$-\text{Cov}[X,Y] - \text{E}[X]\text{E}[Y] + \text{E}[X]\text{E}[Y] - b_1(\text{E}[X])^2 + b_1\text{Var}[X] + b_1(\text{E}[X])^2= 0$
$\Rightarrow -\text{Cov}[X,Y] + b_1\text{Var}[X] = 0$
$\Rightarrow b_1 =  \frac{\text{Cov}[X,Y]}{\text{Var}[X]}$
$\square$




### 2. 

(1) $\text{R}^2$

The function form of $\text{R}^2$ is $\frac{\text{TSS-RSS}}{\text{TSS}}$ and it measures the proportion of variance explained by the model, and ranges from 0 to 1 where 0 indicates that X explains 0% of the variance in Y and 1 indicates 100%. 


(2) Residual Standard Error (RSE):

The functional form of RSE is $\sqrt{\frac{\text{RSS}}{\text{n}-2}}$ and it measures how well the model accounts for your data.

(3) F-Statistic:

The functional form of the F-statistic is 
$\frac{(\text{TSS - RSS})\frac{1}{\text{p}}}{\text{RSS}\frac{1}{\text{(n-p-1)}}}$ and it tells you whether the variance not explained by your model i.e, TSS is greater than the variance explained by your model i.e., RSS. Higher F-statistic indicates greater variability.

```r
#attach(cbpp)
plot(size, incidence, main = "Relationship between herd size and CBPP incidence")
```

#### (b)

```r
fe.fit = lm(incidence~size, data = cbpp)
summary(fe.fit)
```

With a one unit increase in the herd size there is 0.1387 times increase in incidence , i.e., with a one unit increase in the herd size, the number of new serological cases increases 0.1387 times. Thus, the positive coefficient of herd size indicates that herd size has an increasing impact on disease rate. 

#### (c)

```r
re.fit = lmer(incidence~size + (size|herd), data = cbpp)
summary(re.fit)
```

Upon including the herd identifier as a random effect, we notice that the t-value of the fixed effect reduces from 3.150 to 3.030. This indicates that we are getting a more conservative model fit on the fixed effects. 

#### (d) 

```r
ic = AIC(fe.fit, re.fit)
diff(ic$AIC)
```

Upon comparing the simple linear model with the mixed effect model using AIC, we get a difference of -10.104 which is not a very big difference. This indicates that adding the random effect to the model does not significantly change or improve the fixed effect. 

### 3.10

#### (a)

```r
library(ISLR)
attach(Carseats)

mod1 = lm(Sales ~ Price + Urban + US, data = Carseats)
summary(mod1)
```

#### (b)

The negative coefficient of `Price` indicates that with a one unit increase in `Price`, the `Sales` decrease -0.054 times, holding all other variables constant. In other words, if the price company charges for car seats increases by one unit, the unit sales at each location decreases 0.054 times. 
The negative coefficient of `Urban` indicates that if the store is in an urban location (Urban = Yes), the sales would be 0.02 times lower compared to stores in the rural location. In other words, as we shift from Urban to Rural location, the Sales decrease 0.02 times. 
The positive coefficient of `US` indicates that if the store is in US, the sales would be 1.2 times greater compared to the sales if the store was not in the US. In other words, as we move from store in the US and store not in US, the Sales increase 1.2 times. 

#### (c)

$\text{Sales} = \hat{\beta}_0 +  \hat{\beta}_1*\text{Price} +  \hat{\beta}_2*\text{ Urban} +  \hat{\beta}_3*\text{ US} + \epsilon$


$\text{Sales} = 13.043 - 0.054*\text{Price} - 0.022*\text{ Urban} + 1.200*\text{ US} + \epsilon$


#### (d)

We notice that US and Price both have significant p-values which are less than the alpha level 0.05. Thus, these two predictors can reject the null hypothesis. 

#### (e)

For the smaller mode, we remove `Urban` and only keep the two signfiicant predictors `Price` and `US`. 

```r
mod2 = lm(Sales ~ Price + US, data = Carseats)
summary(mod2)
```

#### (f)

The second model fits the data slighly better as it has an adjusted R-squared of 0.2354, while the first model has an adjusted R-squared of 0.2335. Thus, the model with lesser predictor variables fits the data better. 

#### (g)

```r
library(knitr)
confint(mod2)
```

#### (h)

```r
par(mfrow=c(2,2))
plot(mod2)
```

The plots seem to indicate a fairly linear relationship with no explicity outliers, however, we notice that one of the points has a higher leverage than the other points (~0.05).

### 3.13

#### (a)


```r
set.seed(1)
x = rnorm(100, mean=0, sd=1)
```

#### (b)

```r
eps =  rnorm(100, mean=0, sd=sqrt(0.25))
```

#### (c)

```r
y = -1 + (0.5)*x + eps
length(y)
```

The length of the vector `y` is 100. The $\beta_0$ value is -1 and the $\beta_1$ value is 0.5 

#### (d)

```r
plot(x,y, main = "Relationship between x and y")
```

From the scaterplot we notice that x and y have a positive linear relationship. 

#### (e) 

```r
lm.fit = lm(y~x)
summary(lm.fit)
```

Upon fitting a least square model to predict `y` using `x` we obtain the value of $\hat{\beta}_0$ as -1.01885 and the value of $\hat{\beta}_1$ as 0.49947. These values are approximately the same as the $\beta_0$ and $\beta_1$ values of our model.

#### (f)

```r
plot(x,y, main = "Relationship between x and y")
abline(lm.fit, col = "red")
abline(-1,0.5, col = "blue")
legend("bottomright", c("Least Squares Line","Population Regression Line"),col = c("red", "blue"), lty = c(1,1))
```

#### (g)

```r
lm.fit.1 = lm(y ~ x + I(x^2))
summary(lm.fit.1)
```

Upon adding the quadratic term we notice that there is a very slight increase in the adjusted R-squared value from 0.4619 to 0.4672. However, we notice that the quadratic term has a p-value of 0.164 which is not significant, therefore, we do not have sufficient evidence that the quadratic term improves the model fit. 

#### (h)

```r
set.seed(1)
x = rnorm(100, mean=0, sd=1)
eps =  rnorm(100, mean=0, sd=0.25)
y = -1 + (0.5)*x + eps
length(y)

lm.fit.2 = lm(y~x)
summary(lm.fit.2)

plot(x,y, main = "Relationship between x and y")
abline(lm.fit, col = "red")
abline(-1,0.5, col = "blue")
legend("bottomright", c("Least Squares Line","Population Regression Line"),col = c("red", "blue"), lty = c(1,1))
```

In the scatterplot for our new model, we reduced the variance to 0.0625. As a result of this, in the scatter plot above we can clearly say the reduction in noise due to the smaller variance. Additionally, upon fitting a least square model to predict `y` using `x` we obtain the value of $\hat{\beta}_0$ as -1.00942 and the value of $\hat{\beta}_1$ as  0.49973. These values are approximately the same as the $\beta_0$ and $\beta_1$ values of our model.
